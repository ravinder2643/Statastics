#outliers using Zscore

```python
dataset =[11,10,12,14,15,14,13,103,12,14,17,19,1007,13,12,15,16,18,19,202,12,10,12,10,9,21,12,18,16,15]
```

```python
outliers=[]

def detect_outliers(data):
    threshold=3  ## 3rd standard deviation
    mean= np.mean(data)
    std= np.std(data)

    for i in data:
        z_score= (i-mean)/std
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers
            
    
```

```python
plt.hist(dataset)
```

```python
detect_outliers(dataset)
```

IQR 

1. sort data 
2. calculate Q! and Q3
3. find (Q3-Q1)
4. find lower fence and upper fence

```python
dataset = sorted(dataset)
dataset
```

```python
q1, q3 = np.percentile(dataset,[25,75])
```

```python
print(q1)

print(q3)
iqr=q3-q1
```

```python
# lower fence and higher fence

lower_fence= q1-(1.5*iqr)
higher_fence= q3=(1.5*iqr)
```

```python
print(lower_fence,higher_fence)
```
